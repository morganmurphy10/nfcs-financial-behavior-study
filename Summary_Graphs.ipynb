```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Model 1 Graphs

### Exhibit 1: Advisory Type Distribution by Age Group Among U.S. Investors (2024 NFCS Data)

```python
plt.rcParams["font.family"] = "Times New Roman"

age_model1 = pd.DataFrame({
    "year": [
        2015, 2015, 2015,
        2018, 2018, 2018,
        2021, 2021, 2021,
        2024, 2024, 2024
    ],
    "age_group": [
        "18-34", "35-54", "55+",
        "18-34", "35-54", "55+",
        "18-34", "35-54", "55+",
        "18-34", "35-54", "55+"
    ],
    "hybrid": [
        16.30, 9.39, 4.20,
        59.05, 40.44, 19.81,
        45.36, 35.09, 17.50,
        39.71, 33.43, 23.33
    ],
    "robo": [
        33.48, 15.23, 2.60,
        30.00, 40.22, 35.74,
        48.93, 52.56, 44.59,
        53.43, 53.76, 40.77
    ],
    "traditional": [
        50.22, 75.38, 93.20,
        10.95, 19.33, 44.44,
        5.71, 12.35, 37.91,
        6.86, 12.81, 35.90
    ]
})

year_to_plot = 2024

plot_df = (
    age_model1[age_model1["year"] == year_to_plot]
    .set_index("age_group")[["traditional", "hybrid", "robo"]]
)

ax = plot_df.plot(
    kind="bar",
    stacked=True,
    figsize=(9, 6),
    edgecolor="black"
)

for i in range(len(plot_df.index)):
    
    plt.text(i, plot_df["traditional"].iloc[i] / 2,
             f"{plot_df['traditional'].iloc[i]:.1f}%",
             ha='center', va='center', fontsize=10, color='white')
    
    plt.text(i, plot_df["traditional"].iloc[i] + plot_df["hybrid"].iloc[i] / 2,
             f"{plot_df['hybrid'].iloc[i]:.1f}%",
             ha='center', va='center', fontsize=10, color='white')
    
    plt.text(i,
             plot_df["traditional"].iloc[i] + plot_df["hybrid"].iloc[i] + plot_df["robo"].iloc[i] / 2,
             f"{plot_df['robo'].iloc[i]:.1f}%",
             ha='center', va='center', fontsize=10, color='white')
    

plt.title(f"Model 1: Advisory Type by Age Group ({year_to_plot})", fontsize=15)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Percent of Respondents", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Advisory Type", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

### Exhibit 2: Growth and Stabilization of Robo-Advisory Adoption Among U.S. Investors (2015–2024 NFCS Data)

```python
robo_trend = pd.DataFrame({
    "year": [2015, 2018, 2021, 2024],
    "robo_percent": [11.74, 29.21, 38.95, 38.38]
})

plt.figure(figsize=(8, 5))
plt.plot(
    robo_trend["year"],
    robo_trend["robo_percent"],
    marker="o",
    linewidth=2
)

for x, y in zip(robo_trend["year"], robo_trend["robo_percent"]):
    plt.text(x, y + 1, f"{y:.1f}%", ha="center", fontsize=10)

plt.title("Model 1: Robo-Advisory Adoption Rate Over Time", fontsize=15)
plt.xlabel("NFCS Wave", fontsize=12)
plt.ylabel("Robo Users (% of All Cases)", fontsize=12)
plt.xticks(robo_trend["year"])
plt.ylim(0, 45)
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

## Model 2 Graphs

### Exhibit 3: Comparison of Financial Planning Behavior by Advisory Type Using Weighted Planning Index Scores (2015–2024 NFCS Data)

```python
plt.rcParams["font.family"] = "Times New Roman"

planning_index = pd.DataFrame({
    "year": [2015, 2018, 2021, 2024],
    "traditional": [-0.0575, -0.0511, 0.3254, 0.3061],
    "hybrid": [0.3296, 0.1297, -0.1691, -0.1489],
    "robo": [0.1782, -0.0361, -0.2827, -0.2305]
})


x = np.arange(len(planning_index["year"]))
width = 0.25

plt.figure(figsize=(10, 6))

plt.bar(x - width, planning_index["traditional"], width, label="Traditional", edgecolor="black")
plt.bar(x, planning_index["hybrid"], width, label="Hybrid", edgecolor="black")
plt.bar(x + width, planning_index["robo"], width, label="Robo", edgecolor="black")

for i in range(len(planning_index)):
    plt.text(x[i] - width, planning_index["traditional"].iloc[i],
             f"{planning_index['traditional'].iloc[i]:.2f}",
             ha="center", va="bottom" if planning_index["traditional"].iloc[i] >= 0 else "top", fontsize=9)

    plt.text(x[i], planning_index["hybrid"].iloc[i],
             f"{planning_index['hybrid'].iloc[i]:.2f}",
             ha="center", va="bottom" if planning_index["hybrid"].iloc[i] >= 0 else "top", fontsize=9)

    plt.text(x[i] + width, planning_index["robo"].iloc[i],
             f"{planning_index['robo'].iloc[i]:.2f}",
             ha="center", va="bottom" if planning_index["robo"].iloc[i] >= 0 else "top", fontsize=9)

plt.axhline(0, color="black", linewidth=1)
plt.title("Model 2: Planning Index by Advisory Type and Year", fontsize=15)
plt.xlabel("NFCS Wave", fontsize=12)
plt.ylabel("Weighted Planning Index", fontsize=12)
plt.xticks(x, planning_index["year"])
plt.legend(title="Advisory Type")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

### Exhibit 4: Longitudinal Trends in Financial Planning Behavior Across Advisory Types (2015–2024 NFCS Data)

```python
plt.figure(figsize=(10, 6))

plt.plot(planning_index["year"], planning_index["traditional"],
         marker="o", linewidth=2, label="Traditional")

plt.plot(planning_index["year"], planning_index["hybrid"],
         marker="o", linewidth=2, label="Hybrid")

plt.plot(planning_index["year"], planning_index["robo"],
         marker="o", linewidth=2, label="Robo")

for col in ["traditional", "hybrid", "robo"]:
    for x_val, y_val in zip(planning_index["year"], planning_index[col]):
        plt.text(x_val, y_val + 0.025,
                 f"{y_val:.2f}",
                 ha="center", fontsize=9)

plt.axhline(0, color="black", linewidth=1)
plt.title("Model 2: Planning Index Trends by Advisory Type", fontsize=15)
plt.xlabel("NFCS Wave", fontsize=12)
plt.ylabel("Weighted Planning Index", fontsize=12)
plt.xticks(planning_index["year"])
plt.legend(title="Advisory Type")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

## Model 3 Graphs

```python
plt.rcParams["font.family"] = "Times New Roman"

bri = pd.DataFrame({
    "year": [2015, 2018, 2021, 2024],
    "traditional": [-0.0247, -0.1493, -0.0987, 0.0753],
    "hybrid": [0.2684, 0.3028, 0.6820, 0.4319],
    "robo": [0.4920, -0.1211, -0.0041, -0.0246]
})

components_2024 = pd.DataFrame({
    "advisory_type": ["traditional", "hybrid", "robo"],
    "literacy_score": [3.7989, 3.7121, 4.2467],
    "bias_vulnerability_index": [0.4329, 0.5969, 0.4615],
    "overconfidence_gap": [-0.0093, 0.6867, -0.1910],
    "BRI": [0.0753, 0.4319, -0.0246]
})
```

### Exhibit 5: Differences in Bounded Rationality Across Advisory Types Using Weighted BRI Scores (2015–2024 NFCS Data)

```python
x = np.arange(len(bri["year"]))
width = 0.25

plt.figure(figsize=(10, 6))

plt.bar(x - width, bri["traditional"], width, label="Traditional", edgecolor="black")
plt.bar(x, bri["hybrid"], width, label="Hybrid", edgecolor="black")
plt.bar(x + width, bri["robo"], width, label="Robo", edgecolor="black")

# Value labels
for i in range(len(bri)):
    for offset, col in zip([-width, 0, width], ["traditional", "hybrid", "robo"]):
        y = bri[col].iloc[i]
        plt.text(
            x[i] + offset,
            y + 0.025 if y >= 0 else y - 0.025,
            f"{y:.2f}",
            ha="center",
            va="bottom" if y >= 0 else "top",
            fontsize=9
        )

plt.axhline(0, color="black", linewidth=1)
plt.title("Model 3: Bounded Rationality Index by Advisory Type and Year", fontsize=15)
plt.xlabel("NFCS Wave", fontsize=12)
plt.ylabel("Weighted BRI", fontsize=12)
plt.xticks(x, bri["year"])
plt.legend(title="Advisory Type")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

### Exhibit 6: Longitudinal Trends in BRI Index Across Advisory Types (2015–2024 NFCS Data)

```python
plt.figure(figsize=(10, 6))

for col in ["traditional", "hybrid", "robo"]:
    plt.plot(
        bri["year"],
        bri[col],
        marker="o",
        linewidth=2,
        label=col.title()
    )

    for x_val, y_val in zip(bri["year"], bri[col]):
        plt.text(
            x_val,
            y_val + 0.03 if y_val >= 0 else y_val - 0.03,
            f"{y_val:.2f}",
            ha="center",
            va="bottom" if y_val >= 0 else "top",
            fontsize=9
        )

plt.axhline(0, color="black", linewidth=1)
plt.title("Model 3: BRI Trends by Advisory Type", fontsize=15)
plt.xlabel("NFCS Wave", fontsize=12)
plt.ylabel("Weighted BRI", fontsize=12)
plt.xticks(bri["year"])
plt.legend(title="Advisory Type")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

### Exhibit 7: Breakdown of the Bounded Rationality Index with Literacy, Bias Vulnerability, and Overconfidence Components by Advisory Type (2024 NFCS Data)

```python
plt.rcParams["font.family"] = "Times New Roman"

components_2024 = pd.DataFrame({
    "advisory_type": ["Traditional", "Hybrid", "Robo"],
    "Literacy Score": [3.7989, 3.7121, 4.2467],
    "Bias Vulnerability": [0.4329, 0.5969, 0.4615],
    "Overconfidence Gap": [-0.0093, 0.6867, -0.1910],
    "Final BRI": [0.0753, 0.4319, -0.0246]
})

x = np.arange(len(components_2024["advisory_type"]))
width = 0.22

plt.figure(figsize=(10, 6))

plt.bar(x - width, components_2024["Bias Vulnerability"],
        width, label="Bias Vulnerability", edgecolor="black")

plt.bar(x, components_2024["Overconfidence Gap"],
        width, label="Overconfidence Gap", edgecolor="black")

plt.bar(x + width, components_2024["Literacy Score"],
        width, label="Literacy Score", edgecolor="black", alpha=0.45)

plt.scatter(x, components_2024["Final BRI"],
            color="black", s=90, label="Final BRI", zorder=5)

# Add value labels for BRI dots
for i, value in enumerate(components_2024["Final BRI"]):
    plt.text(
        x[i],
        value + 0.08 if value >= 0 else value - 0.12,
        f"BRI: {value:.2f}",
        ha="center",
        va="bottom" if value >= 0 else "top",
        fontsize=10,
        fontweight="bold"
    )

plt.axhline(0, color="black", linewidth=1)
plt.title("Model 3: What Drives the Bounded Rationality Index? (2024)", fontsize=15)
plt.xlabel("Advisory Type", fontsize=12)
plt.ylabel("Reported Weighted Mean", fontsize=12)
plt.xticks(x, components_2024["advisory_type"])
plt.legend(title="Measure", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```
